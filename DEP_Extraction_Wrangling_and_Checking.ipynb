```python
import pandas as pd

# Load the Excel file into a Pandas DataFrame
file_path = "/content/Population_Dataset.xlsx"  # Update with your file path
sheet_names = ['Table 1', 'Table 2', 'Table 3']

# Initialize an empty list to store extracted dataframes
extracted_dataframes = []

for sheet_name in sheet_names:
    # Read the specific sheet into a DataFrame, skipping the first 7 rows
    df = pd.read_excel(file_path, sheet_name=sheet_name, skiprows=7)

    # Filter out rows where the 'Year' column contains non-numeric values
    df = df[pd.to_numeric(df['Year'], errors='coerce').notna()]

    # Convert 'Year' column to integers
    df['Year'] = df['Year'].astype(int)

    # Filter for years 2011 to 2021 and SA3 names 'Melbourne City' or 'Port Philip'
    filtered_df = df[(df['Year'].between(2011, 2021)) &
                     (df['SA3 name'].isin(['Melbourne City', 'Port Phillip']))]

    # Select required columns
    selected_columns = ['Year', 'SA3 name', 'SA2 code', 'SA2 name'] + list(filtered_df.columns[11:])
    filtered_df = filtered_df[selected_columns]

    # Append the filtered DataFrame to the list
    extracted_dataframes.append(filtered_df)

# Display the final DataFrame
print(extracted_dataframes)
```

```python
# Read the CSV file into a pandas DataFrame
tree_df = pd.read_csv('/content/trees-with-species-and-dimensions-urban-forest.csv')

# Convert 'Year Planted' column to string type
tree_df['Year Planted'] = tree_df['Year Planted'].astype(str)

# Extract the last 4 digits from the 'Year Planted' column
tree_df['Year Planted'] = tree_df['Year Planted'].str[-4:]

print(tree_df.shape)
```

```python
# Convert 'Date Planted' to datetime format
tree_df['Date Planted'] = pd.to_datetime(tree_df['Date Planted'])

# Extract the year portion of the 'Date Planted' column
tree_df['Date_Planted_Year'] = tree_df['Date Planted'].dt.year

# Initialize a counter for incorrect year count
incorrect_year_count = 0

# Iterate over each row in the DataFrame
for index, row in tree_df.iterrows():
    # Extract year from 'Date_Planted_Year' column
    year_planted_year = row['Date_Planted_Year']
    # Extract year from 'Year Planted' column
    year_planted = row['Year Planted']

    # Check if 'Year Planted' value is not equal to the year value from 'Date_Planted_Year'
    if int(year_planted) != year_planted_year:
        incorrect_year_count += 1

# Print the count of rows where 'Year Planted' value is not equal to the year value from 'Date_Planted_Year'
print("Number of rows where 'Year Planted' value is not equal to the year value from 'Date_Planted_Year':", incorrect_year_count)

```

```python
tree_df['Year Planted'] = tree_df['Year Planted'].astype('Int32')

# Filter rows where 'Year Planted' values match 'Date_Planted_Year' values
tree_df = tree_df[tree_df['Year Planted'] == tree_df['Date_Planted_Year']]

# Reset index after filtering
tree_df.reset_index(drop=True, inplace=True)

# Display the filtered DataFrame
print(tree_df.shape)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the website
url = "https://8billiontrees.com/carbon-offsets-credits/carbon-ecological-footprint-calculators/how-much-carbon-does-a-tree-capture/#:~:text=How%20Much%20Carbon%20Does%20a%20Tree%20Offset%3F,it%20reaches%2040%20years%20old.&text=An%20acre%20of%20mature%20trees,drive%20your%20car%2026%2C000%20miles."

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content
    soup = BeautifulSoup(response.text, "html.parser")

    # Find the table containing tree species and carbon emissions data
    table = soup.find("table")

    # Check if the table exists
    if table:
        # Extract table rows
        rows = []
        for row in table.find_all("tr")[1:]:
            # Extract data from each row
            cells = row.find_all("td")
            if len(cells) == 2:
                rows.append([cell.text.strip() for cell in cells])

        # Create a pandas DataFrame from the extracted data
        scrapped_df = pd.DataFrame(rows, columns=["Tree Species", "Carbon Emissions Captured in Pounds of CO2"])

        # Print the DataFrame
        print(scrapped_df)
    else:
        print("No table found on the webpage.")
else:
    print("Failed to retrieve the webpage. Status code:", response.status_code)
```

```python
# Remove 'trees' or 'species' from the entries in the 'Tree Species' column
scrapped_df['Tree Species'] = scrapped_df['Tree Species'].str.replace('trees', '', case=False)
scrapped_df['Tree Species'] = scrapped_df['Tree Species'].str.replace('species', '', case=False)

# Print the modified DataFrame
scrapped_df
```

```python
import re
# Define a function to calculate the average of two numbers
def calculate_average(value):
    # If the value is already a float, return it unchanged
    if isinstance(value, float):
        return value
    # Remove commas from the value if present
    value = value.replace(',', '')
    # Check if the value contains 'to' to indicate a range
    if isinstance(value, str) and 'to' in value:
        # Split the value based on the 'to' separator and extract the numbers
        limits = re.findall(r'\d+', value)
        lower_limit = int(limits[0])
        upper_limit = int(limits[1])
        print(lower_limit, upper_limit)
        # Calculate the average of the two limits
        average = (lower_limit + upper_limit) / 2
        return average
    else:
        # If the value doesn't contain 'to', convert it to an integer
        return float(value)
```

```python
print(calculate_average(scrapped_df['Carbon Emissions Captured in Pounds of CO2'][23]))
```

```python
print(scrapped_df['Carbon Emissions Captured in Pounds of CO2'][23])
```

```python
# Apply the calculate_average function to the 'Carbon Emissions Captured in Pounds of CO2' column
scrapped_df['Carbon Emissions Captured in Pounds of CO2'] = scrapped_df['Carbon Emissions Captured in Pounds of CO2'].apply(lambda x: calculate_average(x))

# Print the modified DataFrame
print(scrapped_df)
```

```python
import pandas as pd

# Create DataFrame from dictionaries
tree_df = pd.read_csv("/content/ext_cln_tree_data.csv")
co2_df = pd.read_csv("/content/scrapped_tree_CO2_data.csv")
"""

tree_df = pd.DataFrame({
    'CoM ID': [1013382, 1013383, 1013384],
    'Common Name': ['Huntingdon Elm', 'English Elm', 'Dutch Elm'],
    'Useful Life Expectency Value': [10, 20, 10]
})

co2_df = pd.DataFrame({
    'Tree Species': ['Coastal Redwood', 'English Elm', 'Dutch Elm', 'Mangrove'],
    'Carbon Emissions Captured in Pounds of CO2': [500000, 38000, 14400, 700]
})
"""
# Calculate average 'Useful Life Expectency Value'
avg_useful_life = tree_df['Useful Life Expectency Value'].mean()

# Function to perform partial string matching
def partial_string_match(x, y):
    if isinstance(x, str) and isinstance(y, str):
        return y.lower().strip() in x.lower().strip()
    return False

# Filter co2_df based on partial string match
matched_species = []
for name in tree_df['Common Name']:
    matched = co2_df[co2_df['Tree Species'].apply(lambda species: partial_string_match(name, species))]
    if not matched.empty:
        matched_species.append(matched['Tree Species'].iloc[0])
    else:
        matched_species.append(None)
```

```python
# Merge dataframes based on matching conditions
merged_df = pd.merge(tree_df, co2_df,
                      left_on=tree_df['Common Name'].str.lower(),
                      right_on=co2_df['Tree Species'].str.lower().str.strip(), how='left')

# Calculate new column based on specified conditions
avg_useful_life = merged_df['Useful Life Expectency Value'].mean()

merged_df['CO2offset_yr_tonnes'] = merged_df.apply(
    lambda row: row['Carbon Emissions Captured in Pounds of CO2'] / row['Useful Life Expectency Value'] * 0.00045
    if pd.notnull(row['Useful Life Expectency Value'])
    else 150 / avg_useful_life * 0.00045,
    axis=1
)

# Calculate the replacement value
replacement_value = 150 / avg_useful_life * 0.00045

# Replace NaN values in 'CO2offset_yr_tonnes' with the calculated value
merged_df['CO2offset_yr_tonnes'].fillna(replacement_value, inplace=True)

# Display the final DataFrame with the new column
print(merged_df[['CoM ID', 'Common Name', 'Useful Life Expectency Value', 'Carbon Emissions Captured in Pounds of CO2', 'CO2offset_yr_tonnes']])
```
